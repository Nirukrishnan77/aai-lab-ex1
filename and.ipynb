```python
inputs = [(0,0), (0,1), (1,0), (1,1)]
learning_rate = 0.1
def train_gate_step_by_step(name, targets, w1=-0.5, w2=0.7, b=0.9, epochs=5):
 print(f"\n=== {name} Gate Training ===")
 print(f"Initial weights: w1={w1:.3f}, w2={w2:.3f}, b={b:.3f}")
 for epoch in range(1, epochs+1):
 total_error = 0
 print(f"\nEpoch {epoch}:")
 for (x1, x2), target in zip(inputs, targets):
 net = x1*w1 + x2*w2 + b
 y = 1 if net >= 0 else 0
 error = target - y
 print(f"  Input=({x1},{x2}), Target={target} | "
 f"net={net:.3f}, y={y}, error={error:+d}")
 w1 += learning_rate * error * x1
 w2 += learning_rate * error * x2
 b  += learning_rate * error
if error != 0:
 print(f"    Update -> w1={w1:.3f}, w2={w2:.3f}, b={b:.3f}")
 total_error += abs(error)
 print(f"  Epoch {epoch} total error: {total_error}")
if total_error == 0:
 print("  Converged: No errors this epoch.")
  break
 print(f"\nFinal {name} weights: w1={w1:.3f}, w2={w2:.3f}, b={b:.3f}")
 return w1, w2, b
or_targets = [0, 1, 1, 1]
w1_or, w2_or, b_or = train_gate_step_by_step("OR", or_targets)
and_targets = [0, 0, 0, 1]
w1_and, w2_and, b_and = train_gate_step_by_step("AND", and_targets)
```
